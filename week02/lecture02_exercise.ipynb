## Setup — Run this cell first


```python
import pandas as pd
import numpy as np

# World Happiness Report 2023 — representative data
# Source: https://www.kaggle.com/datasets/ajaypalsinghlo/world-happiness-report-2023

df = pd.read_csv('../data/world_happiness_2023.csv')
df.columns = ['Country','Region','Happiness_Score','GDP','Social_Support',
              'Life_Expectancy','Freedom','Generosity','Corruption']


print(f"Dataset: {len(df)} countries, {len(df.columns)} columns")
print(df.head())
```

```python
import plotly.express as px
import plotly.graph_objects as go

# Explore the dataset before you start
print("Regions in dataset:")
print(df['Region'].value_counts())
print("\nScore range:", df['Happiness_Score'].min(), "–", df['Happiness_Score'].max())
print("\nBottom 10 countries:")
print(df.nsmallest(10, 'Happiness_Score')[['Country','Region','Happiness_Score']])
```

## Task 1 — Regional Comparison Bar Chart

**What to build:** A horizontal bar chart showing the **average happiness score by region**, sorted from highest to lowest.

**Requirements:**
- Horizontal orientation (category names are long)
- Sorted by score, descending (so the happiest region is at the top)
- Zero baseline on x-axis
- At least one design choice that goes beyond the Plotly default (colour, annotation, labels, etc.)
- An insight title that answers: *which region stands out and why does it matter?*

**Hint:** Use `df.groupby('Region')['Happiness_Score'].mean()` to compute the averages.


```python
# Step 1: Sort descending
region_avg = (
    df.groupby('Region')['Happiness_Score']
    .mean()
    .reset_index()
    .sort_values('Happiness_Score', ascending=False)
)

# Step 2: Create chart (single color)
fig1 = px.bar(
    region_avg,
    x='Happiness_Score',
    y='Region',
    orientation='h',
    title='Western Europe stands out as the happiest region due to strong economic and social support systems',
    labels={
        'Happiness_Score': 'Average Happiness Score',
        'Region': 'Region'
    }
)

# Apply a single clean color (no gradient)
fig1.update_traces(marker_color='#4C78A8')  # nice professional blue

# Step 3: Improve layout
fig1.update_layout(
    template='simple_white',
    showlegend=False,   # remove legend
    height=600,
    
    xaxis=dict(
        showline=True,
        linewidth=2,
        linecolor='black',
        showgrid=True,
        gridcolor='lightgray',
        zeroline=True,
        zerolinecolor='black'
    ),
    
    yaxis=dict(
        showline=True,
        linewidth=2,
        linecolor='black',
        autorange='reversed'  # highest at top
    )
)

fig1.show()
```

## Task 2 — Bottom vs. Top: A Contrast Story

**What to build:** A bar chart that highlights the **gap between the happiest and least happy countries**, focusing on a specific insight.

**Requirements:**
- Show the **top 8 AND bottom 8 countries** together (16 bars total)
- Use **colour** to distinguish the two groups (not Plotly's default rainbow)
- Add a **visual separator or annotation** that emphasises the gap
- Insight title that tells the story of the gap

**Hint:** Use `pd.concat([df.nlargest(8,'Happiness_Score'), df.nsmallest(8,'Happiness_Score')])` to get both groups.

**Stretch goal:** Add a vertical reference line showing the global average.


```python
# Task 2: Top 8 vs. Bottom 8 contrast
# ------------------------------------

# Step 1: Get top and bottom countries
top8 = df.nlargest(8, 'Happiness_Score').copy()
top8['Group'] = 'Top 8'
bottom8 = df.nsmallest(8, 'Happiness_Score').copy()
bottom8['Group'] = 'Bottom 8'

combined = pd.concat([bottom8, top8]).sort_values('Happiness_Score')
global_avg = df['Happiness_Score'].mean()
print(f"Global average: {global_avg:.2f}")

# Step 2: Build your chart
# YOUR CODE HERE


# Step 2: Create bar chart
fig2 = px.bar(
    combined,
    x='Happiness_Score',
    y='Country',
    orientation='h',
    title='A stark global divide: top countries are nearly twice as happy as the lowest-ranked nations',
    labels={
        'Happiness_Score': 'Happiness Score',
        'Country': 'Country'
    },
    color='Group',
    color_discrete_map={
        'Top 8': '#4C78A8',     # blue
        'Bottom 8': '#E15759'   # red
    }
)

# Remove legend (colors are self-explanatory)
fig2.update_layout(
    template='simple_white',
    showlegend=False,
    height=700,

    xaxis=dict(
        showline=True,
        linewidth=2,
        linecolor='black',
        showgrid=True,
        gridcolor='lightgray',
        zeroline=True
    ),

    yaxis=dict(
        showline=True,
        linewidth=2,
        linecolor='black'
    )
)

# Add separator line between groups
fig2.add_hline(
    y=7.5,  # between 8 bottom + 8 top bars
    line_dash="dash",
    line_color="black",
    opacity=0.6
)

# Add annotation for the gap
fig2.add_annotation(
    x=global_avg,
    y=7.5,
    text="Global average",
    showarrow=True,
    arrowhead=2
)

# Add vertical line for global average (stretch goal)
fig2.add_vline(
    x=global_avg,
    line_dash="dot",
    line_color="black"
)

# Optional: show values on bars
fig2.update_traces(texttemplate='%{x:.2f}', textposition='outside')

fig2.show()
```

## Done? Stretch Goal

If you finish both tasks with time to spare, try this:

**Task 3 (stretch):** Build a **grouped bar chart** comparing 2 sub-factors (e.g. `GDP_per_capita` and `Freedom`) across the 5 most populated regions. Use colour meaningfully and write an insight title.

Regions to include: `'Western Europe'`, `'Latin America'`, `'East Asia'`, `'Sub-Saharan Africa'`, `'South Asia'`


```python
# Step 1: Filter regions
regions = [
    'Western Europe',
    'Latin America and Caribbean',
    'East Asia',
    'Sub-Saharan Africa',
    'South Asia'
]

df_filtered = df[df['Region'].isin(regions)]

# Step 2: Aggregate
agg = df_filtered.groupby('Region')[['GDP', 'Freedom']].mean().reset_index()

# Step 3: Melt
melted = agg.melt(
    id_vars='Region',
    value_vars=['GDP', 'Freedom'],
    var_name='Metric',
    value_name='Value'
)

# Step 4: Ensure order (GDP first, then Freedom)
melted['Metric'] = pd.Categorical(
    melted['Metric'],
    categories=['GDP', 'Freedom'],
    ordered=True
)

# Step 5: Sort within each metric (descending)
melted = melted.sort_values(['Metric', 'Value'], ascending=[True, False])

# Step 6: Plot (GROUPED, not stacked)
fig3 = px.bar(
    melted,
    x='Value',
    y='Region',                 # <-- KEY FIX (NOT Metric)
    color='Metric',             # only 2 colors
    orientation='h',
    barmode='group',
    title='GDP varies far more across regions than freedom, highlighting economic inequality',
    labels={
        'Value': 'Average Score',
        'Region': 'Region',
        'Metric': 'Factor'
    },
    color_discrete_map={
        'GDP': '#4C78A8',
        'Freedom': '#F28E2B'
    }
)

fig3.update_layout(
    template='simple_white',
    height=600,
    showlegend=False,
    
    xaxis=dict(
        showline=True,
        linewidth=2,
        linecolor='black',
        showgrid=True,
        gridcolor='lightgray'
    ),
    
    yaxis=dict(
        showline=True,
        linewidth=2,
        linecolor='black'
    )
)

fig3.update_traces(
    texttemplate='%{x:.2f}',
    textposition='outside'
)

fig3.show()
```
